```python
import pandas as pd
import numpy as np
```

```python
dfEmp = pd.read_csv(r"C:\Stack overflow\Data-Warehouse-Training\CapStone Project\Retail Sales Performance Dashboard\Week-2\employees.csv")
dfPro = pd.read_csv(r"C:\Stack overflow\Data-Warehouse-Training\CapStone Project\Retail Sales Performance Dashboard\Week-2\products.csv")
dfSales = pd.read_csv(r"C:\Stack overflow\Data-Warehouse-Training\CapStone Project\Retail Sales Performance Dashboard\Week-2\sales.csv")
dfStor = pd.read_csv(r"C:\Stack overflow\Data-Warehouse-Training\CapStone Project\Retail Sales Performance Dashboard\Week-2\stores.csv")
```

Checking for null values

```python
print(f"Null values: {dfEmp.isna().sum().any()}")
print(f"Null values: {dfPro.isna().sum().any()}")
print(f"Null values: {dfSales.isna().sum().any()}")
print(f"Null values: {dfStor.isna().sum().any()}")
```

Using Joins

```python
df = dfSales.merge(dfPro, how="inner", on="productID").merge(dfStor, how="inner", on="storeID").merge(dfPro, how="inner", on="productID")
```

```python
df
```

```python
df["revenue"] = df["quantity"] * df["price_y"]
df["profit"] = df["revenue"] - df["cost_y"]
```

```python
df
```

```python
summary = df.groupby("name_y").agg(
    {
        "revenue": "sum",
        "profit": "sum"
    }
)
```

```python
summary
```

Deliverable :- Cleaned dataset, Summarizing key metrics

```python
# 1. cleaned dataset with calculated fields
summary.to_csv("summary.csv")
```

```python
# 2. summarizing key metrics - totalRevenue, totalProfit, topSellingProduct, underSellingProduct, averageSaleValue

totalRevenue = summary["revenue"].sum()
totalProfit = summary["profit"].sum()

productSummary = df.groupby("name_x").agg(
    {
        "quantity": "sum",
        "profit": "sum"
    }
    ).sort_values("profit", ascending=False)

storeSummary = df.groupby("name_y").agg(
    {
        "revenue": "sum",
        "profit": "sum"
    }
).sort_values("profit", ascending=False)

topProduct = productSummary.iloc[0, :]
bottomProduct = productSummary.iloc[-1, :]

topStore = storeSummary.iloc[0, :]
bottomStore = storeSummary.iloc[-1, :]

report = pd.DataFrame(
    {
        "totalRevenue": totalRevenue,
        "totalProfit": totalProfit,
        "topProduct": topProduct.name,
        "bottomProduct": bottomProduct.name,
        "topStore": topStore.name,
        "bottomStore": bottomStore.name
    },
    index=[0]
    )
```

```python
report.to_csv("report.csv")
```
